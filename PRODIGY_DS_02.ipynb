```python
import pandas as pd


titanic_df = pd.read_csv('C:/Users/HP/Desktop/train.csv')
```

```python

titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

titanic_df['HasCabin'] = titanic_df['Cabin'].notnull().astype(int)
titanic_df.drop('Cabin', axis=1, inplace=True)


titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)


titanic_df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.countplot(data=titanic_df, x='Survived')
plt.title('Survival Count')
plt.show()


survival_rate = titanic_df['Survived'].mean() * 100
print(f'Survival Rate: {survival_rate:.2f}%')
```

```python

sns.countplot(data=titanic_df, x='Survived', hue='Sex')
plt.title('Survival Count by Gender')
plt.show()
```

```python

sns.histplot(data=titanic_df, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.show()


sns.histplot(data=titanic_df, x='Age', hue='Survived', bins=30, kde=True)
plt.title('Age Distribution by Survival')
plt.show()
```
